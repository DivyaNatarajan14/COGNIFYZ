REVIEWING THE DATASET

```python
rdata=pd.read_csv('Dataset .csv')
```

SHALLOW COPYING

```python
df=rdata.copy()
```

```python
table_booking_percentage = (df['Has Table booking'].value_counts(normalize=True) * 100).round(2)
print("Percentage of Restaurants Offering Table Booking:")
print(table_booking_percentage)

# Determine the percentage of restaurants that offer online delivery
online_delivery_percentage = (df['Has Online delivery'].value_counts(normalize=True) * 100).round(2)
print("\nPercentage of Restaurants Offering Online Delivery:")
print(online_delivery_percentage)
```

```python
# Plot a pie chart for table booking
plt.figure(figsize=(8, 8))
plt.pie(table_booking_percentage, labels=table_booking_percentage.index, autopct='%1.1f%%', colors=['lightcoral', 'lightgreen'])
plt.title('Percentage of Restaurants Offering Table Booking')
plt.show()


# Plot a pie chart for online delivery
plt.figure(figsize=(8, 8))
plt.pie(online_delivery_percentage, labels=online_delivery_percentage.index, autopct='%1.1f%%', colors=['skyblue', 'gold'])
plt.title('Percentage of Restaurants Offering Online Delivery')
plt.show()
```

```python
# Convert 'Aggregate rating' column to numeric (if not already numeric)
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')

# Compare the average ratings of restaurants with and without table booking
average_rating_table_booking = df.groupby('Has Table booking')['Aggregate rating'].mean().round(2)
print("Average Ratings:")
print(average_rating_table_booking)

# Plot a bar chart to visualize the comparison
plt.figure(figsize=(8, 6))
average_rating_table_booking.plot(kind='bar', color=['lightcoral', 'lightgreen'])
plt.title('Average Ratings Comparison (With and Without Table Booking)')
plt.xlabel('Table Booking')
plt.ylabel('Average Rating')
plt.xticks(rotation=0)
plt.show()
```

```python
# Check unique values in the 'Price range' column
print("Unique Price:", df['Price range'].unique())

# Analyze the availability of online delivery among different price ranges
plt.figure(figsize=(10, 6))
sb.countplot(x='Price range', hue='Has Online delivery', data=df, palette='Set2')
plt.title('Online Delivery Availability over Price Ranges')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```
