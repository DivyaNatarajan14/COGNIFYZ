BAR CHART FOR DESCRIPTIVE STATISTICS

```python
for column in df.select_dtypes(include='number').columns:

# Bar chart for descriptive statistics
  summary_stats = df[column].describe()
  plt.figure(figsize=(5,3))
  sb.barplot(x=summary_stats.index, y=summary_stats.values, palette='viridis')
  plt.title(f'Descriptive Statistics of {column}')
  plt.ylabel('Value')
  plt.xticks(rotation=45,ha='right')
  plt.show()
```

```python
# Explore the distribution of "Country Code"
country_distribution = df['Country Code'].value_counts()
print("Distribution of Country Codes:")
print(country_distribution)

# Plot the distribution of "Country Code"
plt.figure(figsize=(10, 6))
country_distribution.plot(kind='bar', color='skyblue')
plt.title('Distribution of Country Codes')
plt.xlabel('Country Code')
plt.ylabel('Count')
plt.show()
```

```python
# Explore the distribution of "City"
city_distribution = df['City'].value_counts()
print("\nDistribution of Cities:")
print(city_distribution)

# Find values with top 10 occurrences in 'City'
top_10 = city_distribution.head(10)

# Create bar chart to visualize top 10 values
plt.figure(figsize=(15, 8))
top_10.plot(kind='bar', color='skyblue')
plt.title('Top 10 Cities')
plt.xlabel('City')
plt.ylabel('Count')
plt.show()
```

```python
# Explore the distribution of "Cuisines"
cuisine_distribution = df['Cuisines'].value_counts()
print("\nDistribution of Cuisines:")
print(cuisine_distribution.head(10))  # Display the top 10 most common cuisines

# Plot the distribution of "Cuisines"
plt.figure(figsize=(14, 6))
cuisine_distribution.head(10).plot(kind='bar', color='mediumseagreen')
plt.title('Top 10 Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.show()
```
