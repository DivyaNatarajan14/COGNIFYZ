```python
columns_to_skip = ['Restaurant Name', 'City', 'Address', 'Locality', 'Locality Verbose', 'Cuisines', 'Currency']

# Loop through object-type columns
for column in df.select_dtypes(include='object').columns:
    # Skip plotting specified columns
    if column in columns_to_skip:
        continue

    # Bar chart
    plt.figure(figsize=(10, 5))
    sb.countplot(x=df[column])
    plt.title(f'Bar Chart of {column}')
    plt.xticks(rotation=45, ha='right')  # ha='right' means horizontal alignment of text in right side
    plt.show()
```

NUMERICAL FEATURES

HISTOGRAM

```python
for column in df.select_dtypes(include='number').columns:
# Histogram
  plt.figure(figsize=(10,5))
  sb.histplot(df[column], kde=True, color='skyblue')#kde=True adds kernel density estimate plot
  plt.title(f'Histogram of {column}')
  plt.xlabel(column)
  plt.ylabel('Frequency')
  plt.show()
```

BAR CHART FOR DESCRIPTIVE STATISTICS

```python
for column in df.select_dtypes(include='number').columns:
# Bar chart for descriptive statistics
  summary_stats = df[column].describe()
  plt.figure(figsize=(8,4))
  sb.barplot(x=summary_stats.index, y=summary_stats.values, palette='viridis')
  plt.title(f'Descriptive Statistics of {column}')
  plt.ylabel('Value')
  plt.xticks(rotation=45,ha='right')
  plt.show()
```

```python
columns_to_skip = ['Restaurant Name', 'City', 'Address', 'Locality', 'Locality Verbose', 'Cuisines', 'Currency']

# Loop through object-type columns
for column in df.select_dtypes(include='object').columns:
    # Skip plotting specified columns
    if column in columns_to_skip:
        continue


#find values with top 10 occurrences in 'CarName'
plt.figure(figsize=(15,6))
top_10 = (df[column].value_counts()).iloc[:3]

#create bar chart to visualize top 10 values
top_10.plot(kind='bar')
```

```python
columns_to_skip = ['Restaurant Name', 'City', 'Address', 'Locality', 'Locality Verbose', 'Cuisines', 'Currency']

# Loop through object-type columns
for column in df.select_dtypes(include='object').columns:
    # Skip plotting specified columns
    if column in columns_to_skip:
        continue

    sum_values = df[column].sum()

    # Create a bar plot
    plt.figure(figsize=(8,5))
    plt.xticks(rotation=35,ha='right',fontsize=7)

    # Use seaborn's barplot to plot the mean 'Aggregate rating' for each category in the selected numeric column
    sb.barplot(data=df, x=column, y='Aggregate rating', palette='Set2', order=df.groupby(column)['Aggregate rating'].mean().sort_values().index)

    plt.title(f'{column} vs Aggregate rating')
    plt.show()
```
