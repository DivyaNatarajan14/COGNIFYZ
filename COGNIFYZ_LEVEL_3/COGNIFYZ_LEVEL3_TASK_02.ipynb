# **CUSTOMER PREFERNCE ANALYSIS**

CUISINE INTEREST

```python
# Assuming your DataFrame is named 'df'
cuisine_counts = df['Cuisines'].value_counts().head(10)  # Display top 10 cuisines
plt.figure(figsize=(12, 6))
sb.barplot(x=cuisine_counts.values, y=cuisine_counts.index, palette='viridis')
plt.title('Top 10 Cuisines Preferred by Customers')
plt.xlabel('Number of Restaurants')
plt.ylabel('Cuisine')
plt.show()
```

PRICE RANGE INTEREST

```python
plt.figure(figsize=(10, 6))
sb.histplot(df['Price range'], bins=range(1, 7), kde=False, color='skyblue')
plt.title('Distribution of Restaurants by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.show()
```

Rating Analysis

```python
plt.figure(figsize=(10, 6))
sb.histplot(df['Aggregate rating'], bins=30, kde=True, color='salmon')
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Number of Restaurants')
plt.show()
```

ONLINE DELIVERY AND TABLE BOOKING

```python
plt.figure(figsize=(10, 6))
sb.countplot(x='Has Online delivery', data=df, palette='coolwarm')
plt.title('Number of Restaurants Offering Online Delivery')
plt.xlabel('Has Online Delivery')
plt.ylabel('Number of Restaurants')
plt.show()
```

Analyze the relationship between the type of
cuisine and the restaurant's rating.

```python
# Calculate average ratings for each cuisine
avg_ratings_by_cuisine = df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)

# Select the top 10 cuisines
top_25_cuisines = avg_ratings_by_cuisine.head(25)

# Create a horizontal bar plot
plt.figure(figsize=(12, 8))
sb.barplot(x=top_25_cuisines.values, y=top_25_cuisines.index, palette='viridis')
plt.title('Top 20 Cuisines by Average Rating')
plt.xlabel('Average Rating')
plt.ylabel('Cuisine')
plt.show()
```


Identify the most popular cuisines among
customers based on the number of votes.

```python
# Calculate total votes for each cuisine
total_votes_by_cuisine = df.groupby('Cuisines')['Votes'].sum().sort_values(ascending=False)

# Select the top cuisines (e.g., top 10)
top_cuisines = total_votes_by_cuisine.head(10)

# Create a horizontal bar plot
plt.figure(figsize=(12, 8))
sb.barplot(x=top_cuisines.values, y=top_cuisines.index, palette='viridis')
plt.title('Top Cuisines by Total Votes')
plt.xlabel('Total Votes')
plt.ylabel('Cuisine')
plt.show()
```

```python
# Calculate total votes for each locality
total_votes_by_locality = df.groupby('Locality')['Votes'].sum().sort_values(ascending=False)

# Select the top localities (e.g., top 10)
top_localities = total_votes_by_locality.head(10)

# Create a horizontal bar plot
plt.figure(figsize=(12, 8))
sb.barplot(x=top_localities.values, y=top_localities.index, palette='viridis')
plt.title('Top 10 Localities by Total Votes')
plt.xlabel('Total Votes')
plt.ylabel('Locality')
plt.show()
```

Determine if there are any specific cuisines
that tend to receive higher rating

```python
# Calculate average ratings for each cuisine
avg_ratings_by_cuisine = df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)

# Select the top cuisines (e.g., top 30)
top_cuisines = avg_ratings_by_cuisine.head(30)

# Create a horizontal bar plot
plt.figure(figsize=(12, 8))
sb.barplot(x=top_cuisines.values, y=top_cuisines.index, palette='viridis')
plt.title('Top Cuisines by Average Rating')
plt.xlabel('Average Rating')
plt.ylabel('Cuisine')
plt.show()
```

```python
# Filter the DataFrame for restaurants with a rating of 5
df_top_ratings = df[df['Aggregate rating'] == 4.8]

# Create a count plot
plt.figure(figsize=(14, 8))
sb.countplot(x='Cuisines', data=df_top_ratings, palette='viridis')
plt.title('Top Ratings (4.80) vs Cuisine')
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.show()
```
